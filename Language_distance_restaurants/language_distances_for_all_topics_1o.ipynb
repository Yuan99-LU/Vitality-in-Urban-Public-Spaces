```python
import pandas as pd
```

```python
import numpy as np
```

```python
def cos_similar(a,b):  
# manually compute cosine similarity
    dot = np.dot(a, b)
    norma = np.linalg.norm(a)
    normb = np.linalg.norm(b)
    cos = dot / (norma * normb)
    
    return cos
```

```python
for N in range(1, 11):
    data = pd.read_csv('vectors_topic_by_languages_top20_features_each_language_all_top10_topics/vectors_topic_'+
                           str(N)+'_by_languages.csv')
    languages = data.columns.tolist()[3:]
    pd_data = pd.DataFrame({"languages": languages})
    for lang in languages:
        lang_cos = []
        for lang_another in languages:
            freq1 = np.array(data[lang])
            freq2 = np.array(data[lang_another])
            result = cos_similar(freq1, freq2)
            print('%s vs %s: %f' %(lang, lang_another,result))
            #print(result)
            lang_cos.append(result)
        pd_data[lang] = lang_cos
    pd_data.to_csv('language_distance/restaurants_top20_features_each_language_Topic'+ str(N)+'.csv')
```

```python
import pandas as pd
import os
```

```python
LANGUAGES = ['[Arabic]','[Chinese, (Sim.)]','[Chinese, (Trad.)]','[Danish]','[Dutch]','[English]', '[Finnish]',
'[French]',
'[German]',
'[Greek]',
'[Italian]',
'[Japanese]',
'[Norwegian]',
'[Polish]',
'[Portuguese]',
'[Russian]',
'[Serbian]',
'[Spanish]',
'[Swedish]',
'[Turkish]',
'[]']
```

```python
filenames = os.listdir('LDA_by_languages/restaurants_lda_after_pyLDAvis')
```

```python
N1 = 1
N2 = 2

#top_N_group = []
#N=1
for N in range(6, 11):
    top_N_group = []
    for lang in LANGUAGES:
        file_name = 'output_'+lang+'_'+str(N2)+".csv"
        print(lang)
        #data1 = pd.read_csv('output_'+month+'_'+str(N1)+".csv")
        if file_name in filenames:
            data2 = pd.read_csv('LDA_by_languages/restaurants_lda_after_pyLDAvis/'+file_name)
            data2 = data2[data2['Category']=='Topic'+str(N)]
        #data2 = data2.reset_index(drop = True, inplace = True)
            top1_data2 = data2.drop(['term','Category','Total','loglift','logprob'], axis =1)
            top1_data2 = top1_data2.rename(index=str, columns={"Freq": lang+"_Freq", "Term": lang + "_Term"})
            top1_data2 = top1_data2.reset_index(drop = True)
            top_N_group.append(top1_data2)
            print(len(top1_data2))
    combined = pd.concat(top_N_group, axis =1)
    combined.to_csv("combine_all_languages_by_Topic/combined_Topic_"+str(N)+"_all_languages.csv")
```

```python
import pandas as pd
```

```python
data_author = pd.read_csv('"urban vitality".csv', names = 'abcdef')
```

```python
years = [str(x) for x in range(1998, 2019)]
```

```python
years
```

```python
data_author_1998_2008 = data_author[data_author['a']. isin(years)]
```

```python
data_author = data_author_1998_2008
```

```python
data_author
```

```python
data_new2 = pd.DataFrame(data_author['d'].str.split('-',2).tolist(),
                                   columns = ['author','publisher','e-address'])
```

```python
names = []
for name in data_new2['author']:
    #name = str(name).replace(r'\xa0','')
    #print(name)
    names.extend(name.split(','))
#print(names)

```

```python
data_pd = pd.DataFrame(names, columns = ['author'])

data_pd['author'] = data_pd['author'].str.replace(r'[\W_]+',' ')
data_pd['author'] = data_pd['author'].str.strip()
data_pd['author'] = data_pd['author'].str.upper()
counts = data_pd['author'].value_counts()

counts.to_csv('author_counts_in_urban_vitality_1998_2008’.csv')
```

```python
import pandas as pd
import os
```

```python
LANGUAGES = ['[Chinese, (Sim.)]','[Chinese, (Trad.)]','[Czech]','[Danish]','[Dutch]','[English]', '[Finnish]',
'[French]',
'[German]',
'[Greek]',
'[Hungarian]',
'[Indonesian]',
'[Italian]',
'[Japanese]',
'[Norwegian]',
'[Polish]',
'[Portuguese]',
'[Russian]',
'[Serbian]',
'[Spanish]',
'[Swedish]',
'[Thai]',
'[Turkish]']
```

```python
features_topic1 = pd.read_csv('top20_each_language_features_all_top5_topics/Topic_'+'all_top5'+'_features.csv', 
                                  names = ['index_original','features'])
filenames = os.listdir('LDA_by_languages/attractions_lda_after_pyLDAvis')
```

```python
N2 = 2
LANGUAGES_NEW = []
for lang in LANGUAGES:
    file_name = 'output_'+lang+'_'+str(N2)+".csv"
    if file_name in filenames:
        LANGUAGES_NEW.append(lang)
print(LANGUAGES_NEW)
```

```python
for N in range(1, 6):
    topic1 = pd.read_csv('combine_all_languages_by_Topic/combined_Topic_'+str(N)+'_all_languages.csv')
    
    for lang in LANGUAGES_NEW:
        freq = []
        for term in features_topic1['features']:
            #print(term)
            if term in topic1[lang + '_Term'].tolist()[0:20]:
                index = topic1[lang + '_Term'][topic1[lang + '_Term'] == term].index[0]
                freq.append(topic1[lang + '_Freq'][index])
            else:
                freq.append(0)
        print(freq)
        features_topic1[lang + '_Freq'] = freq
    features_topic1.to_csv('vectors_topic_by_languages_top20_features_each_language_all_top5_topics/vectors_topic_'+
                           str(N)+'_by_languages.csv')
```

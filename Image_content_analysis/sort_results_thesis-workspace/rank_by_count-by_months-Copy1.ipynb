```python
import pandas as pd
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
for month in months:
    data = pd.read_csv('by_months/'+month+"_Lund_Photos.csv")
    print(len(data))
    print(data.head(5))
    new = [data['c_x'],data['d_x'],data['e'],data['f'],
           data['g'],data['h'],data['i'], data['j'], data['k']]
    data['string'] = data['b_x'].str.cat(new, sep = ',')
    print(data.head(5))
    #data2= data.drop(['a','b_y','c_y','d_y'], axis =1)
    #count = pd.value_counts(data2.values.flatten())
    data.to_csv('by_months/to_string/' + month +"_combine_strings.csv")
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
for i in range(2003, 2019):
    year = data[data['d_y'].str.contains(str(i))== True]
    print(len(year))
    year.to_csv('by_years/'+str(i)+"_Lund_Photos.csv",index =False)
```

```python
for month in months:
    mon = data[data['d_y'].str.contains(month)== True]
    print(len(mon))
    mon.to_csv('by_months/'+month+"_Lund_Photos.csv",index =False)
```

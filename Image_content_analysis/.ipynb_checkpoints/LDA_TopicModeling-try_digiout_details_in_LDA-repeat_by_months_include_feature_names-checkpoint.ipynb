```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import NMF, LatentDirichletAllocation
import pyLDAvis
import pyLDAvis.sklearn
import pyLDAvis.gensim
import numpy as np
from datetime import datetime, date, time
import csv
import sys
import importlib
import pandas as pd
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
def tf_vctorize(n_max_features, text):
    #n_features = 1000
    tf_vectorizer = CountVectorizer(strip_accents = 'unicode',
                                max_features=n_max_features,
                                stop_words='english',
                                max_df = 0.95,
                                min_df = 2)
    tf = tf_vectorizer.fit_transform(text)
    tf_feature_names = tf_vectorizer.get_feature_names()
    return tf, tf_vectorizer, tf_feature_names
    

def LDA_analyze(tf_vector, n_topics):
    lda = LatentDirichletAllocation(n_topics=n_topics,doc_topic_prior =1/(n_topics*10), max_iter=50,
                                learning_method='online',
                                learning_offset=50.,
                                evaluate_every = 1,
                                random_state=0)
    lda.fit(tf_vector)
    print(lda)
    return lda

def tf_idf(n_max_features, text):
    tfidf_vectorizer = TfidfVectorizer(max_df=0.95,
                                       min_df=2,
                                       max_features=n_max_features,
                                       stop_words='english')
    tfidf = tfidf_vectorizer.fit_transform(text)
    tfidf_feature_names = tfidf_vectorizer.get_feature_names()
    return tfidf, tfidf_feature_names
```

```python
n_max_features=3000
n_topics = 10
n_top_words = 10
```

```python
for month in months:
    data = pd.read_csv('by_months/to_string/'+month+'_combine_strings.csv')
    #print(data['string'].head(1))
    data['string'] = data['string'].replace(np.nan, '', regex=True)
    data['string'] = data['string'].apply(str)
    text_all = (data['string']).tolist()
    print(len(text_all))
    
    tf_vector, tf_vectorizer, tf_feature_names = tf_vctorize(n_max_features, text_all)
    lda = LDA_analyze(tf_vector, n_topics)
    #tf_feature_names_array = np.array(tf_feature_names)
    MyFile=open("LDA_by_months/LDA_original/output_"+ month+ "_orgi_lda_feature_names"  +'.csv', 'w')

    for element in tf_feature_names:
        MyFile.write(element)
        MyFile.write(',')
    MyFile.write('\n')
    
    for feature_list in lda.components_:
        for elem in feature_list:
            MyFile.write(str(elem))
            MyFile.write(',')
        MyFile.write('\n')
        
    MyFile.close()
    
    #np.savetxt("LDA_by_months/LDA_original/output_"+ month+ "_orgi_lda"  +'.csv', tf_feature_names_array, delimiter=",")
    #np.savetxt("LDA_by_months/LDA_original/output_"+ month+ "_orgi_lda"  +'.csv', lda.components_, delimiter=",")
    
    #pd.read_csv()

'''
    lda_prepared = pyLDAvis.sklearn.prepare(lda, tf_vector, tf_vectorizer)

    N=0
    for i in lda_prepared[0:3]:
        #i.shape()
        #print(i)
        N+=1
        i.to_csv("LDA_by_months/output_"+ month+ "_" + str(N) +'.csv')
'''    
```

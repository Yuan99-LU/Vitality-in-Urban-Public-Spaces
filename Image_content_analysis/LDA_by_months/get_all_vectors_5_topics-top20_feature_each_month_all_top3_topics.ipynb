```python
import pandas as pd
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
for N in range(1, 6):
    topic1 = pd.read_csv('Combine_all_months_by_Topic/combined_Topic_'+str(N)+'_all_months.csv')
    features_topic1 = pd.read_csv('top20_each_month_features_all_top3_topics/Topic_'+'all_top3'+'_features.csv', 
                                  names = ['index_original','features'])
    for month in months:
        freq = []
        for term in features_topic1['features']:
            #print(term)
            if term in topic1[month + '_Term'].tolist()[0:20]:
                index = topic1[month + '_Term'][topic1[month + '_Term'] == term].index[0]
                freq.append(topic1[month + '_Freq'][index])
            else:
                freq.append(0)
        print(freq)
        features_topic1[month + '_Freq'] = freq
    features_topic1.to_csv('vectors_topic_by_months_top20_features_each_month_all_top3_topics/vectors_topic_'+str(N)+'_by_months.csv')
```

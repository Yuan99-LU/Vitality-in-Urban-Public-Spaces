```python
import pandas as pd
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
N1 = 1
N2 = 2

for N in range(1, 6):
    
    features_pre = []
    for month in months:
        print(month)
        data2 = pd.read_csv('LDA_PyLDAvis/output_'+month+'_'+str(N2)+".csv")
        data2 = data2[data2['Category']=='Topic'+str(N)]
        #data2 = data2.reset_index(drop = True, inplace = True)
        features_pre.append(data2['Term'])
        #print(features_pre)
    features  = pd.concat(features_pre).drop_duplicates().reset_index(drop=True)
    print(features)
    features.to_csv('features_by_topics/Topic_'+str(N)+'_features.csv')
   
```
